```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cc.csv", sep=",", encoding = "UTF-8")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(), inplace = True)
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(), inplace = True)
df.isnull().sum()
```

```python
df.drop(columns = ['CUST_ID'], inplace = True)
df.head()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 

model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)

```

```python
import matplotlib.pyplot as plt

agglomerative=pd.DataFrame(clust_labels)

fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ['ONEOFF_PURCHASES'] , df ["PURCHASES"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("One off Purchases")
ax.set_ylabel("Purchases")

plt.colorbar(scatter)
```

```python
import scipy.cluster.hierarchy as shc
plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('sample index')
plt.ylabel('distance')
shc.dendrogram(shc.linkage(df, method = 'complete'),
    truncate_mode='lastp',  
    p=25,  
    show_leaf_counts=False,  
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  
)
plt.show()
```

# Finding the k best value

```python
from sklearn.cluster import KMeans 
SSD=[]
K = range(2, 10)
for k in K :
    km = KMeans(n_clusters = k)
    km = km.fit(df)
    SSD.append(km.inertia_)
plt.plot(K, SSD, 'bx-')
plt.xlabel('k')
plt.ylabel('SSD')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
km = KMeans(n_clusters = 3)
km = km.fit(df)
labels=pd.DataFrame(km.labels_)
df['label'] = labels
df.head()
```

```python


sns.scatterplot(data=df, x='ONEOFF_PURCHASES', y='PURCHASES', hue='label')
plt.title('Distribution of clusters')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
datad = pca.fit_transform(df)

sil = []

for k in K:
    km = KMeans(n_clusters = k).fit(df)
    labels = km.labels_
    sil.append(silhouette_score(datad, labels, metric = 'euclidean'))

plt.plot(K, sil, "bo-")
plt.title('Silhoutte scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutte score')
plt.show()
```

# If we go by the silhouette score, we should chose k = 2

```python
km = KMeans(n_clusters = 2)
km = km.fit(df)
labels=pd.DataFrame(km.labels_)
df['label'] = labels

sns.scatterplot(data=df, x='ONEOFF_PURCHASES', y='PURCHASES', hue='label')
plt.title('Distribution of clusters')
plt.show()
```

# It seems  k = 3 clusters is more accurate and the kmeans method seems more precise than the agglomerative method for clustering customers
